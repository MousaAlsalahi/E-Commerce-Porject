###  Cross-Vsalidation Video 4

  * instead of splitting data into 2 split one for test and another for train we can split it to more folds  to make the result more reliable
  * so each split has multiple folds(cut) one fold for test and the rest for train and calculate the metric R_Squared of this split
  * then for the second split change the fold of test and the rest for fold train get the split R_Squared
  * at the end we have kfold num R_squared for each by which we can get the mean median standard deviation 95% confidence
  * 5folds = 5fold cv cross validation
  * kfolds = kfold cv

  ************
  * steps
      * import cross_val and Kfold (set a seed(random) and shuffle)from sklearn.model_selection
      * assign kf =Kfold()
            * shuffle to switch data before splitting
            * random to make get same result each time so it will sbe split the same if repeated the process
            * n_splits number of folds default 5
      * instantiate the model like reg,knn
      * cross va score
      * now we have the result which is array for each split
      * now we can calculate the mean to see the average
      * std to see how each split CV changes across the splits or folds
          * A std of 0.1  noticeable variation
          * A std of 0.01  very stable
          * A std of 0.001  extremely stable
      * the 95% interval
          * np.quantile(cv_results, [0.025, 0.975])
          * shows you the best-case and worst-case performance your model reaches across all cross-validation folds.
          * to understand how much your models performance can change.

```python
from sklearn.model_selection import cross_val_score , Kfold
kf = Kfold( n_splits=6, shuffle=True, random_state=42)
reg=LinearRegression()
cv_result = cross_val_score(reg,x,y,cv=kf)

np.mean(cv_result),np.std(cv_result)

np.quantile(cv_result,[0.025,0.975])
# containing the upper and lower limit of intervals
```
