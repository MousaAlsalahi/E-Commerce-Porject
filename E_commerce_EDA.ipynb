##Bivariate Analysis: Explore relationships between two variables

#### Order Status & Order IDs

```python
# Aggregate order IDs by order status to see frequency of each status
orders_per_status_count=(
            orders_df.groupby("order_status",as_index=False)["order_id"]
            .count()
            .sort_values("order_id",ascending=False)
)

```

```python

sns.barplot(
    data=orders_per_status_count,
    x="order_status",
    y="order_id",
    hue="order_status",
    palette="viridis"
    )
plt.title('Number of Orders by Status')
plt.xlabel('Order Status')
plt.ylabel('Count of Orders')
plt.xticks(rotation=45)
plt.show()
```

* Observation: The vast majority of orders were successfully delivered.

####Order ID & Purchase Timestamp

```python
# Count the number of orders per month to analyze monthly trends
orders_count_per_month=(
             orders_df.groupby("purchase_month",as_index=False)
             .agg(n_orders=("order_id","count"))
             )

sns.lineplot(
    data=orders_count_per_month,
    x="purchase_month",
    y="n_orders",
    )
plt.title("Number of Orders by Month",y=1.05)
plt.ylabel("Number of Orders")
plt.xticks(range(1,13),labels,rotation=45)
plt.xlabel("Months")
plt.show()
```

* Observation: There is a sharp decrease in orders in September, followed by a slight increase in the subsequent months.

####Order Approved Timestamp & Order Purchase Timestamp

```python
# Calculate waiting time between order purchase and approval
orders_df["approval_waiting_time_hours"]=orders_df["order_approved_at"]-orders_df["order_purchase_timestamp"]
# Convert waiting time to days and hours for easier interpretation
orders_df["approval_waiting_time_hours"]=(orders_df["approval_waiting_time_hours"].dt.total_seconds() / 3600).round(1)
orders_df["approval_waiting_time_hours"]


```

```python
monthly_waiting_time=(
                orders_df.groupby("purchase_month",as_index=False)
                .agg({"approval_waiting_time_hours":"sum"})

)
```

```python
sns.barplot(
    data=monthly_waiting_time,
    x="purchase_month",
    y="approval_waiting_time_hours",
    palette="viridis"
    )

plt.title('Duration of Waiting Time Monthly')
plt.xlabel('Month')
plt.ylabel('Approval Waiting Time')
plt.xticks(range(12),labels,rotation=45)
plt.show()
```

* Observation: Waiting time noticeably increases during the second and third quarters of the year, followed by a slight decrease in the last quarter.

####Order Purchase Timestamp & Customer Delivery Timestamp

```python
# Calculate time difference between order purchase and delivery in days
orders_df["purchase_to_customer_delivery_days"]= orders_df["order_delivered_customer_date"]\
                              -orders_df["order_purchase_timestamp"]

orders_df["purchase_to_customer_delivery_days"]= (orders_df["purchase_to_customer_delivery_days"].dt.total_seconds() / (24*3600)).round(1)

days_to_delivery_per_month=(
                      orders_df.groupby("purchase_month",as_index=False)
                      .agg({"purchase_to_customer_delivery_days":"median"})
)
```

```python
# Visualize average delivery time (in days) from purchase to customer by month
sns.lineplot(
    data=days_to_delivery_per_month,
    x="purchase_month",
    y="purchase_to_customer_delivery_days"
    )

plt.xlabel("Months")
plt.ylabel("Number of Days to Delivery")
plt.xticks(range(1,13),labels)
plt.show()
```

* Observation:
- February has the highest average purchase-to-delivery time, which decreases steadily until August before rising again.

* At this point, you have both the mean review score and mean delivery speed for every category across all months.

3. extract the highest-rated category in each month to identify the monthly leaders in customer satisfaction.

4. To put it all together, visualize the monthly mean review score and delivery time—using dual axes—so the relationship between faster delivery and higher satisfaction becomes clear on a single timeline.
